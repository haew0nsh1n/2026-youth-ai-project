# APIM 경유 Azure OpenAI 호출 테스트

```python
from openai import AzureOpenAI

APIM_ENDPOINT = "https://apim-ai-workshop-010.azure-api.net"
SUBSCRIPTION_KEY = "166e6c71e7e843e8844437a27521face"

client = AzureOpenAI(
    azure_endpoint=APIM_ENDPOINT,
    api_key=SUBSCRIPTION_KEY,
    api_version="2024-12-01-preview",
    default_headers={"Ocp-Apim-Subscription-Key": SUBSCRIPTION_KEY},
)

response = client.chat.completions.create(
    model="gpt-5.4-mini",
    messages=[{"role": "user", "content": "안녕 Azure에 대해 설멸해줘"}],
    max_completion_tokens=2000,
)

print(f"✅ APIM 정상 작동 확인!")
print(f"모델: {response.model}")
print(f"응답: {response.choices[0].message.content}")
```
